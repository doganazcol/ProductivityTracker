```python
import pandas as pd
import matplotlib.pyplot as plt

# Sample DataFrame for illustration
data = {
    'leetcode': [3, 0, 1],
    'research': [0,0, 1],
    'gym': [1, 0, 1],
    'healthy_eating': [1, 0, 1],
    'applied_res_homework': [0,0, 1],
    'histart_res_homework': [1, 0, 1],
    'urban_homework': [0, 0, 1],
    '198_homework': [0, 0, 1],
    'ethics_homework': [0, 0, 1]
}

df = pd.DataFrame(data)

# Plot each metric separately
for column in df.columns:
    plt.figure(figsize=(6, 4))
    plt.title(f'{column} Metric Over Time')

    # Iterate over the values to plot each segment with the correct color
    for i in range(1, len(df[column])):
        color = 'red' if df[column][i] < df[column][i - 1] else 'green'
        plt.plot([i - 1, i], [df[column][i - 1], df[column][i]], color=color, linewidth=2)

    plt.xlabel('Time')
    plt.ylabel('Score')
    plt.grid(True)
    plt.show()
```
